```
def return_count(string):
  count_dict = {}
  count = 1
  for i, char in enumerate(string):
    if i == 0:
      continue
    
    if string[i-1] == char and len(string) == i + 1 and count >= 1:
      count += 1
      if count_dict.get(char, 0) < count:
        count_dict[string[i-1]] = count

    if string[i-1] == char:
      count += 1
    elif count > 1 and count_dict.get(char, 0) < count:
      count_dict[string[i-1]] = count
      count = 1
   
  print(count_dict)


print(return_count(s4))
```

```
testfile = '/content/drive/My Drive/Snap_color.png'
img = cv2.imread(testfile)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
mask1 = cv2.inRange(img, (200, 0, 0), (255, 50, 50))
target = cv2.bitwise_and(img, img, mask=mask1)
plt.imshow(np.array(target).astype('uint8'))
plt.show()

target_bl = cv2.cvtColor(target, cv2.COLOR_RGB2GRAY)
image, contours, hierarchy = cv2.findContours(target_bl,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)


for contour in contours:
  x,y,w,h = cv2.boundingRect(contour)
  if w < h:
    continue
  c_img = cv2.drawContours(img, [contour], -1, (0,255,0), 3)
plt.imshow(np.array(c_img).astype('uint8'))
```
